```python
total_logs =  pd.concat([answerlogs, reclogs], axis=0, join='outer')
```

```python
total_logs = total_logs.sort_values(by=['timestamp'])
```

```python
total_logs = total_logs.groupby('user_id')[['timestamp', 'answer_id', 'rectype_id', 'company_id', 'job_small_id', 'question_type_id', 'impressions']]
```

```python
for name, group in total_logs:
    print(f'name: {name}')
    print(f'group: {group}')
    print()
```
